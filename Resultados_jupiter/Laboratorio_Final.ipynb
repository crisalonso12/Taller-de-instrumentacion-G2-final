### Etapa diferencial y de aislamiento

&nbsp;&nbsp;&nbsp; A continuación, se explicaran cada una de las etapas del circuito mostrado en la figura anterior: 

#### Etapa diferencial:

&nbsp;&nbsp;&nbsp; La etapa diferencial es la encargada de amplificar la tensión de entrada proveniente de una termocupla, termistor u otro dispositivo analógico que convierta de una propiedad física (temperatura, intensidad lumínica, entre otros) a una señal eléctrica.


![diferencial](Diferencial.png)
 


#### Etapa de ganancia y seguidor:

&nbsp;&nbsp;&nbsp; La segunda y tercera etapa están compuestas, primeramente, por una configuración de ganancia, que, con ayuda de un multiplexor, se puede aumentar o disminuir el voltaje proveniente de la primera etapa, dicha etapa de ganancia invierte el voltaje, es decir, pasa de un voltaje positico a uno negativo, por lo que es necesario volver a invertir la tensión, de modo que se añade una etapa con una configuración de seguidor, el cual no afecta a la magnitud del voltaje pero si la fase, de manera que, la salida es un voltaje positivo.


![ganancia](Ganancia.png)


#### Etapa del octocoplador:

&nbsp;&nbsp;&nbsp; Por último, la etapa del octocplador esta compuesta de tres etapas. La primera de estas etapas es una etapa diferencial el cual genera un offset que no aporta mucho voltaje pero si aumenta la corriente, esto con el objetivo de lograr una correcta polarización del diodo del octocoplador 4N35. Seguidamente, hay una etapa con una configuración llamada "bomba de corriente de Howland", que cumple con la función de redireccionar toda la corriente directo a la siguiente etapa, de esta manera se asegura que toda la corriente proveniente de la etapa anterior se dirija al octocoplador.


![octo](Octocoplador.png)
